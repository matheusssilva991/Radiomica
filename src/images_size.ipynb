```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

## Mamografia

```python
df_mamografia = pd.read_csv("../outputs/mamografia/images_size.csv", index_col=0)

fig, ax = plt.subplots(1, 1, figsize=(16, 8))
sns.histplot(ax=ax, x="size_mb", data=df_mamografia, hue="database")
ax.set_title("Distribuio do tamanho das imagens em MegaByte (MB) - Todos Bancos", size=15)
ax.set_ylabel("Quantidade", size=14)
ax.set_xlabel("Tamanho em MegaByte (MB)", size=14)
plt.show()
```

```python
databases = ["DDSM", "CMMD", "INBREAST", "MIAS"]
indices = [(0, 0), (0, 1), (1, 0), (1, 1)]
colors = ['coral', 'cornflowerblue', 'indianred', 'lightgreen']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for database, indice, color in zip(databases, indices, colors):
    i, j = indice
    sns.histplot(ax=axes[i][j], data=df_mamografia[df_mamografia['database'] == database], x='size_mb', 
                 color=color, bins=15)
    axes[i][j].set_title(f"Distribuio do tamanho das imagens em MegaByte (MB) - {database}", size=15)
    axes[i][j].set_ylabel("Quantidade", size=14)
    axes[i][j].set_xlabel("Tamanho em MegaByte (MB)", size=14)

fig.subplots_adjust(bottom=0.1, right=0.9, top=1)
plt.show()
```

## Ultrassom

```python
UBS_metadata = pd.read_csv("../outputs/ultrassom/UBS_metadata.csv")
HMSS_metadata = pd.read_csv("../outputs/ultrassom/HMSS_metadata.csv")
THAMMASAT_metadata = pd.read_csv("../outputs/ultrassom/THAMMASAT_metadata.csv")
BUSI_metadata = pd.read_csv("../outputs/ultrassom/BUSI_metadata.csv")
OASBUD_metadata = pd.read_csv("../outputs/ultrassom/OASBUD_metadata.csv")

metadatas = [UBS_metadata, HMSS_metadata, THAMMASAT_metadata, BUSI_metadata, OASBUD_metadata]
databases_names = ['UBS', 'HMSS', 'THAMMASAT', 'BUSI', 'OASBUD']

merge_metadatas = pd.DataFrame()

for metadata, name in zip(metadatas, databases_names):
    tmp = metadata[['image_size_mb']].copy()
    tmp['database'] = [name] * tmp.shape[0]
    merge_metadatas = pd.concat([merge_metadatas, tmp], axis=0)
```

Todos os bancos juntos

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
sns.histplot(ax=ax, x="image_size_mb", data=merge_metadatas[merge_metadatas['database'] != 'UBS'], hue="database")
ax.set_title("Distribuio do tamanho das imagens em MegaByte (MB) - Todos Bancos", size=15)
ax.set_ylabel("Quantidade", size=14)
ax.set_xlabel("Tamanho em MegaByte (MB)", size=14)
plt.show()
```

```python
databases = ["UBS", "HMSS", "THAMMASAT", "BUSI", "OASBUD"]
indices = [(0, 0), (0, 1), (1, 0), (1, 1), (2, 0)]
colors = ['coral', 'cornflowerblue', 'indianred', 'lightgreen', 'lightpink']

fig, axes = plt.subplots(3, 2, figsize=(17, 17))

for database, indice, color in zip(databases, indices, colors):
    i, j = indice
    sns.histplot(ax=axes[i][j], data=merge_metadatas[merge_metadatas['database'] == database], x='image_size_mb', 
                 color=color, bins=15)
    axes[i][j].set_title(f"Distribuio do tamanho das imagens em MegaByte (MB) - {database}", size=15)
    axes[i][j].set_ylabel("Quantidade", size=14)
    axes[i][j].set_xlabel("Tamanho em MegaByte (MB)", size=14)

fig.subplots_adjust(bottom=0.1, right=0.9, top=1)
plt.show()
```
